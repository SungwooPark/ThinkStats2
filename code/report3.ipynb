```python
import thinkstats2
import pandas
import thinkplot
import numpy as np
import statsmodels.tsa.stattools as smtsa
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

%matplotlib inline
```

```python
col_names = ["credit_score", "first_payment_date", "first_homebuying", "maturity_date", "msa_code", "MI_percentage", "num_units","occp_status", "CLTV", "DTI","UPB", "LTV", "interest_rate","channel", "PPM", "product_type", "state", "property_type", "PO_code", "iden", "loan_purpose", "loan_term", "num_borrowers", "seller_name","service_name","super_conforming_flag"]
df_1999 = pandas.read_csv("report3_data/1999.csv", header=None)
df_1999.drop(df_1999.columns[[26, 27]], axis=1, inplace=True)
df_1999.columns = col_names
```

```python
df_1999
```

```python
df_2006 = pandas.read_csv("report3_data/2006.csv", header=None)
df_2006.drop(df_2006.columns[[26]], axis=1, inplace=True)
df_2006.columns = col_names
df_2006
```

```python
df_2015 = pandas.read_csv("report3_data/2015.csv", header=None)
df_2015.drop(df_2015.columns[[26, 27]], axis=1, inplace=True)
df_2015.columns = col_names
```

```python
type(df_2006["UPB"][0])
#numpy.int64
df_2006["credit_score"] = pandas.to_numeric(df_2006["credit_score"], errors='coerce')
```

```python
#Regression model for 2006 dataset (credit score as a dependent variable)
formula = 'credit_score ~ MI_percentage + CLTV + UPB + LTV + interest_rate'
#formula = 'credit_score ~ UPB'
results = smf.ols(formula, data=df_2006).fit()
results.summary()
```

```python
#Regression model for 2006 dataset (interest rate as dependent variable)
formula = 'interest_rate ~ MI_percentage + CLTV + UPB + LTV + credit_score'
results = smf.ols(formula, data=df_2006).fit()
results.summary()
```

```python
#Regression model for 2015 dataset (credit score as a dependent variable)
formula = 'credit_score ~ MI_percentage + CLTV + UPB + LTV + interest_rate'
results = smf.ols(formula, data=df_2015).fit()
results.summary()
```

```python
#Regression model for 2015 dataset (interest rate as dependent variable)
formula = 'interest_rate ~ MI_percentage + CLTV + UPB + LTV + credit_score'
results = smf.ols(formula, data=df_2015).fit()
results.summary()
```
